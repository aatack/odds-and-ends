```python
from adversarial_image_classification.helpers import paste_image


# This image is of "brass": copy it to the clipboard then run this cell
# https://github.com/EliSchwartz/imagenet-sample-images/blob/master/n02892201_brass.JPEG?raw=true
image = paste_image()
```

```python
# Some "hashes" of the image - used to check whether it's different to similar images
image.shape, image.sum()
```

```python
from adversarial_image_classification.noise_addition import maximise_probability

# Interrupt the process to stop early.  Should take ~300 iterations
results = maximise_probability(image, "fountain", max_iterations=1000)
```

```python
# Throws an error if certain obvious constraints are not satisfied, and renders the
# output image for visual inspection
results.check()
```

```python
from adversarial_image_classification.helpers import render_tensor


print(f"Initial probability: {results.initial_probability:.4f}")
print(f"Final probability: {results.final_probability:.4f}")

render_tensor(results.noise, normalise=True)
```

```python
print(f"Maximum magnitude of noise: {results.noise.abs().max().item():.4f}")
```
